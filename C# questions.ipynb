```python
#Question 3
static double triangleArea(int a , int b, int c)
{
    float s = (a + b + c) / 2;
    double area = Math.Sqrt( s * (s - a) * (s - b) * (s - c));
    return area;
}
double areaTriangle = triangleArea(2,2,2);
Console.WriteLine("Area of Triangle: "+areaTriangle);
```
